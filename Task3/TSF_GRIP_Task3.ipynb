## Exploratory Data Analysis (EDA)


### plot data using pairplot:
Pairplot visualizes given data to find the relationship between them where the variables can be continuous or categorical.

Plot pairwise relationships in a data-set for all of the numeric variables within our dataset.

 we can see that we have all of our variables shown along the y and x axis. Along the diagonal we have a histogram showing the distribution of each of the variables.

```
sns.pairplot(retail_df, kind='reg', plot_kws={'line_kws':{'color':'red'}})
```

From the pair plot we notice:
Profit & Sales is very correlaited

### find correlations using corr()
At application such regression modelling its necessary to remove correlated variables to improve the model.
We can find correlations using corr() function and visualize the correlation matrix using a heatmap in seaborn.

```
plt.figure(figsize = (15, 15))
sns.heatmap(retail_df.corr(),annot=True,cmap='BuPu')
plt.show()
```

We notice that:

"Profit" has strong positive correlation with "Sales" whereas it has strong negative correlation with "Discount"

"Sales" has strong negative correlation with "Discount"
